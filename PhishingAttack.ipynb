```python
import pandas as pd
```

```python
data = pd.read_csv(r"/content/Phishing_Legitimate_full.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data.info()
data.drop("id",axis=1,inplace=True)
```

```python
data.corr()
```

```python
import seaborn  as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
corrmat = data.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(25,25))
g = sns.heatmap(data[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
# dependent features
Y = data["CLASS_LABEL"]

# Independent features
dropping = ["CLASS_LABEL","UrlLengthRT","SubdomainLevelRT","PctExtResourceUrlsRT","AbnormalExtFormActionR","ExtMetaScriptLinkRT","PctExtNullSelfRedirectHyperlinksRT"]
X = data.drop(dropping,axis=1)
```

```python
X.head()
```

```python
Y.head()
```

```python
# Feature important
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
model.fit(X,Y)
```

```python
model.feature_importances_
```

```python
# Plot the feature importance
feat_importances = pd.Series(model.feature_importances_,index=X.columns)
feat_importances.nlargest(10).plot(kind="barh")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
predict = model.predict(X_test)
```

```python
sns.displot(predict)
```

```python
sns.displot(Y_test-predict)
```

```python
model.score(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
import pickle
file = open("Phising_model.pkl","wb")
pickle.dump(model,file)
```
